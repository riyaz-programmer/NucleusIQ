# Text-to-SQL Evaluation Showcase (NucleusIQ)

Five evaluation patterns for an **Autonomous** text-to-SQL agent on Chinook — graded from `AgentResult`.

**Blog:** [Evaluating text-to-SQL agents with NucleusIQ](https://www.nucleusbox.com/evaluating-text-to-sql-agents-with-nucleusiq/)

Run all evals: `python run_all.py`

```python
import asyncio
import sys
from pathlib import Path

PKG = Path.cwd()
if str(PKG) not in sys.path:
    sys.path.insert(0, str(PKG))

from agent.env import load_env
from agent.build_agent import make_llm, build_sql_agent
from nucleusiq.agents.task import Task
from evals.extract import final_answer, tool_names

load_env()
```

```python
db = PKG / "data" / "chinook.sqlite"
if not db.exists():
    from scripts.download_chinook import download
    download()

llm = make_llm()
agent = build_sql_agent(llm)
await agent.initialize()
result = await agent.execute(Task(id="demo", objective="How many customers are from Canada?"))
print("Answer:", final_answer(result))
print("Trajectory:", tool_names(result))
```
